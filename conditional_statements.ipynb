````python
#write a program to find the given number is positive or negative or zero?
number = int(input("please enter number here:"))
if(number>0):
   print(number,"is positive")
elif(number<0):
   print(number,"is negative")
else:
     print(number,"is zero")
````

````python
"""
write a program to find the student grades based on marks
1. if student marks in between 100 to 80 then grade is A 
2. if student marks in between 79 to 60 then grade is B 
3. if student marks iin between 59 to 40 then grade is c
4.if student marks in between 39 to 45 then grade is d
5.if student marks in between 24 to 0 then grade is fail
6.if other than above marks expected output should be please enter the correct marks!
"""
marks=int(input("enter student marks here:"))
if (marks>=80 and marks <=100):
    print(marks,"is A grade")
elif (marks>=60 and marks<=79):
    print(marks,"is B grade")
elif (marks>=40 and marks<=59):
    print(marks,"is c grade")
elif (marks>=25 and marks<=39):
    print(marks,"is D grade")
elif(marks>=0 and marks<=24):
    print(marks,"is Fail")
else: print(marks,"is not correct marks, please enter correct marks")
    
````
