# iterations in python
1. iterations means loops to work with repeatative tasks or repetative performences
2. python have 2 types of iterators
3. `for`loop
4. `while`loop

```python- for loop
for iter_variable in sequence: 
    print("for is keyword to start the for loop")
    print("iter_variables is one type of variable to iterate data from sequence")
    print("in is keyword to validate sequence is present or not")
    print("sequence is collection like range,list,string,tuple,..etc"
```

```python - while loop
while(condition):
    print("while is keyword to start the iteration based on condition")
    print("condition is true then executed inside statements")
```

````python
#write a program to print 1 to 50 numbers using for loop?
for num in range(1,51):
    print(num)
````

````python
#write a program to iterate name using for loop 
name="shivaji"
for char in name:
    print(char)
````

````python
#write a program to find the even numbers in between 100 to 150 using for loop?
for n in range(100,151):
    if (n%2==0):
       print(n,end=",")
````

````python
#write a program to find the odd numbers in between 100 to 150 using for loop ?
for n in range(100,151):
    if (n%2==0):
        print(n,end=",")
````

````python
#write a program to find the vowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char in "aeiouAEIOU"):
        print(char)
````

````python
#write a program to find the non vowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char not in "aeiouAEIOU"):
        print(char)
    
````

````python
#write a program to read name from list using for loop 
names=["vikas","shivaji","aditya","chandra"]
for n in names:
    print("my name is",n)
````
